## Di bawah ini adalah kode lama, sebaiknya biarkan saja

```python
import os

import requests
from requests_html import HTMLSession
import pandas as pd
```

```python
try:
    os.makedirs('SBMPTN/Universitas')
except:
    pass

try:
    os.makedirs('SBMPTN/Jurusan')
except:
    pass
```

```python
namaUniv = []
namaJurusan = []
daya_tampung = []
peminat = []

daftar_univ = HTMLSession()

link = 'https://sidata-ptn.ltmpt.ac.id/ptn_sb.php'

url = daftar_univ.get(link)

html = url.html

univ_univ = html.find('tr')[1:]

for univ in univ_univ:

    nama_univ = univ.find('td')[2].text.split('\n')[0]

    link_jurusan = os.path.join(link,univ.find('td')[3].find('a')[0].attrs['href'])

    daftar_jurusan = HTMLSession()

    url = daftar_jurusan.get(link_jurusan)

    html = url.html

    jurusan_jurusan = html.find('tr')[2:]

    for jurusan in jurusan_jurusan:

        try:
            info_jurusan = jurusan.find('td')
            nama_jur = info_jurusan[2].text.lower()
            daya_tampung_jur = info_jurusan[4].text
            peminat_jur = info_jurusan[5].text
            
            namaUniv.append(nama_univ)
            namaJurusan.append(nama_jur.upper())
            daya_tampung.append(daya_tampung_jur)
            peminat.append(peminat_jur)

            print(nama_univ,nama_jur.upper(),daya_tampung_jur,peminat_jur)
        except:
            pass
```

```python
namaJurusan = [nama.upper() for nama in namaJurusan]

d = {'Universitas':namaUniv,'Jurusan':namaJurusan,'Daya Tampung 2021':daya_tampung,'Peminat 2020':peminat}

df = pd.DataFrame(data=d)
```

```python
df['Daya Tampung 2021'] = df['Daya Tampung 2021'].astype(int)

for i in range(len(df)):
    try:
        df.loc[i,'Peminat 2020'] = int(df.loc[i,'Peminat 2020'])
    except:
        df.loc[i,'Peminat 2020'] = int(df.loc[i,'Peminat 2020'][0] + df.loc[i,'Peminat 2020'][2:])
```

```python
for i in range(len(df)):
    if df.loc[i,'Daya Tampung 2021'] <= df.loc[i,'Peminat 2020']:
        df.loc[i,'Persentase'] = round((df.loc[i,'Daya Tampung 2021']/df.loc[i,'Peminat 2020'])*100,2)
    elif df.loc[i,'Daya Tampung 2021'] > df.loc[i,'Peminat 2020']:
        df.loc[i,'Persentase'] = 100
```

```python
df = df.sort_values('Universitas',ascending=False)
```

```python
df.to_csv('SBMPTN/sbmptn2021.csv',index=False)
df.to_excel('SBMPTN/sbmptn2021.xlsx',index=False)
```

```python
df = pd.read_csv('sbmptn2021.csv')
```

## Universitas
File dikelompokan berdasarkan **Universitas**

```python
for univ in df['Universitas'].unique():
    condition = (df['Universitas'] == univ)
    univ_df = df.loc[condition].sort_values('Jurusan')
    univ_df.to_csv(f'SBMPTN/Universitas/{univ}.csv',index=False)
```

## Jurusan
File dikelompokan berdasarkan **Jurusan**

```python
for jurusan in sorted(df['Jurusan'].unique()):
    if '/' or '.' in jurusan:
        jurusan = jurusan.replace('/','-')
        
    condition = (df['Jurusan'] == jurusan)
    jurusan_df = df.loc[condition].sort_values('Universitas')
    jurusan_df.to_csv(f'SBMPTN/Jurusan/{jurusan}.csv',index=False)
```
